```python
df["Gross"]=df["Gross"]/100000
df["budget"]=df["budget"]/100000
df
```

```python
# Calculating the profit by subtracting budget from gross and adding new column in our data set 
df["profit"]=df["Gross"]-df["budget"]
df
```

```python
df = df.sort_values(by='profit',ascending=False)
df.iloc[:10,:]
```

```python
df["MetaCritic"]=df["MetaCritic"]/10
#movies average ratings
df["Avg_rating"]=(df["IMDb_rating"]+df["MetaCritic"])/2
df
```

```python
# Sorting in descending order of avrage movie rating
df2=df[["Title","IMDb_rating","MetaCritic","Avg_rating"]]
df2=df2.loc[abs(df2["IMDb_rating"]-df2["MetaCritic"]<0.5)]
```

```python
# Creating a dataframe by genre
df_genre=df.loc[:,"CVotes10":"VotesnUS"]
df_genre[["genre_1","genre_2","genre_3"]]=df[["genre_1","genre_2","genre_3"]]
df_genre
```

```python
df.columns
```

```python
df_genre["cnt"]=1
df_genre[["genre_1","genre_2","genre_3"]]
#grouping the movies by individual genre
df_g1=df_genre.groupby("genre_1").aggregate(np.sum)
df_g2=df_genre.groupby("genre_2").aggregate(np.sum)
df_g3=df_genre.groupby("genre_3").aggregate(np.sum)
df_g1
```

```python
df_g2
```

```python
df_g3
```

```python
# adding the grouped data and storing in new data frame
df_add=df_g1.add(df_g2,fill_value=0)
df_add=df_add.add(df_g3,fill_value=0)
df_add
```

```python
genre_top10=df_add.loc[df_add["cnt"]>10]
genre_top10
```

```python
genre_top10.iloc[:,0:-1]=genre_top10.iloc[:,0:-1].divide(genre_top10["cnt"],axis=0)
genre_top10
```

```python
# roundig off the columns of votes
genre_top10.loc[:,"VotesM":"VotesnUS"]=round(genre_top10.loc[:,"VotesM":"VotesnUS"],2)
# Convrting Cvotes to int type
genre_top10[genre_top10.loc[:,"CVotes10":"CVotesnUS"].columns]=genre_top10[genre_top10.loc[:,"CVotes10":"CVotesnUS"].columns].astype(int)
genre_top10
```

```python
# Count plot for genre
sns.barplot(x=genre_top10["cnt"],y=genre_top10.index)
plt.show()
```

```python
# heat map for CVotes-related column
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ax=sns.heatmap(genre_top10[["CVotesU18M","CVotes1829M","CVotes3044M","CVotes45AM"]],annot=True,cmap="coolwarm")
plt.subplot(1,2,2)
ax=sns.heatmap(genre_top10[["CVotesU18F","CVotes1829F","CVotes3044F","CVotes45AF"]],annot=True,cmap="coolwarm")
plt.show()
```
